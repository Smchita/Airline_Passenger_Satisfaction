               3) Analyse bivariée
On cherche à étudier la corrélation entre les variables quantitative continue, parce que dans la suite on en aura besoin pour faire la classification. Alors les variables dépendantes seront réduites pour éviter la redondance des variables corrélé.

```python
with sns.axes_style('white'):
    plt.figure(figsize= (6, 2))
    sns.heatmap(df_clean[variable_continue].corr(), annot = True, fmt = '.2f', linewidths= 0.8, cmap="YlGnBu")

plt.figure(figsize = (4, 2))
sns.scatterplot(x = 'Departure Delay in Minutes', y = 'Arrival Delay in Minutes', data = df_clean, color = 'teal')
plt.show()
```

**Forte corrélation**, **0.97**, entre le **retard de départ** et le **retard 
d'arrivé**, ceci dit que les deux variables apportent 
la même information sur le jeu de donnée donc inutile de 
garder les deux, c'est pourquoi l'un d'entre eux doit être supprimé:

```python
df_clean.drop('Departure Delay in Minutes', axis=1, inplace= True)
```
